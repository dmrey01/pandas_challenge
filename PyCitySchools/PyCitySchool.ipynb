```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path



#File to Load 
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
#Calculate the total number of unique schools
schools_count_df= school_data_complete_df["school_name"].nunique()
schools_count_df
```

```python
#calculate the total number of students
student_count_df =school_data_complete_df["student_name"].nunique()
student_count_df
```

```python
# Calculate the total budget
total_budget_df = school_data_df["budget"].sum()
total_budget_df 
```

```python
#calculate the average (mean) math score 
average_math_score_df =  school_data_complete_df["math_score"].mean()
average_math_score_df
```

```python
#Calculate the average (mean) reading score
average_reading_score_df = school_data_complete_df["reading_score"].mean()
average_reading_score_df
                                               
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count_df = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)].count()["student_name"]
passing_math_percentage_df = passing_math_count_df / float(student_count_df) * 100
passing_math_percentage_df
```

```python
#Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count_df = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage_df = passing_reading_count_df / float(student_count_df) * 100
passing_reading_percentage_df
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count_df = school_data_complete_df[
    (school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate_df = passing_math_reading_count_df /  float(student_count_df) * 100
overall_passing_rate_df
```

```python
#Create a high-level snapshot of the district's key metrics in a DataFrame
#school_data_complete_df.columns
district_summary_df = pd.DataFrame({
    "Total_School": [schools_count_df],
    "Total_Students": [student_count_df],
    "Total_Budget": [total_budget_df],
    "Math_Score": [average_math_score_df],
    "Reading_Score": [average_reading_score_df],
    "Passing_Math_Percentage": [passing_math_percentage_df],
    "Passing_Reading_Percentage": [passing_reading_percentage_df],
    "Overall_Passing_Rate": [overall_passing_rate_df]})
                                              
                       
# Formatting
district_summary_df["Total_Students"] = district_summary_df["Total_Students"].map("{:,}".format)
district_summary_df["Total_Budget"] = district_summary_df["Total_Budget"].map("${:,.2f}".format)
district_summary_df["Reading_Score"] = district_summary_df ["Reading_Score"].map("{:.2f}".format)
district_summary_df["Math_Score"] = district_summary_df["Math_Score"].map("{:.2f}".format)
district_summary_df["Passing_Math_Percentage"] = district_summary_df["Passing_Math_Percentage"].map("{:.1f}%".format)
district_summary_df["Passing_Reading_Percentage"]= district_summary_df["Passing_Reading_Percentage"].map("{:.1f}%".format)
district_summary_df["Overall_Passing_Rate"]= district_summary_df["Overall_Passing_Rate"].map("{:.1f}%".format)
                                                    
# Display the DataFrame, 
district_summary_df.head()
```

```python
# Use the code provided to select all of the school types
school_types_df = school_data_df.set_index(["school_name"])["type"]
school_types_df.head()
```

```python
# Calculate the total student count per school
per_school_counts_df = groupby_schools_df["Student ID"].count()
per_school_counts_df.head()
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget_df = groupby_schools_df["budget"].mean()
#per_school_budget_df = per_school_budget_df.map("{":,}".format)
per_school_budget_df.head()
```

```python
per_school_capita_df = per_school_budget_df / per_school_counts_df
per_school_capita_df.head()
```

```python
# Calculate the average test scores per school
per_school_math_df = groupby_schools_df["math_score"].median()
#per_school_math_df = per_school_math_df.map("{:.0f}%".format)
per_school_math_df.head()
```

```python
per_school_reading_df= groupby_schools_df["reading_score"].median()
#per_school_reading_df= per_school_reading_df.map("{:.0f}%".format)
per_school_reading_df.head()
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math_df = groupby_schools_df["math_score"].apply(lambda x:(x >= 70).sum())
school_students_passing_math_df = students_passing_math_df / per_school_counts_df *100
students_passing_math_df.head()
 
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading_df = groupby_schools_df["reading_score"].apply(lambda x:(x >= 70).sum())
school_students_passing_reading_df =students_passing_reading_df / per_school_counts_df * 100
school_students_passing_reading_df.head()
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
#students_passing_math_and_reading = school_data_complete[
#   (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
#]
#school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
Students_passing_math_and_reading_df = school_data_complete_df[
    (school_data_complete_df["reading_score"] >= 70) & (school_data_complete_df["math_score"] >= 70)
]
school_students_passing_math_and_reading_df = Students_passing_math_and_reading_df.groupby(["school_name"]).size()
school_students_passing_math_and_reading_df.head()
                                                       
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math_df = school_students_passing_math_df / per_school_counts_df * 100
per_school_passing_reading_df = school_students_passing_reading_df / per_school_counts_df * 100
overall_passing_rate_df = school_students_passing_math_and_reading_df / per_school_counts_df * 100
overall_passing_rate_df.head()
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

per_school_summary_df = pd.DataFrame({
    "School Type" : school_types_df,
    "Total Students": per_school_counts_df,
    "Total School Budget": per_school_budget_df,
    "Per Student Budget" :per_school_capita_df,
    "Average Math Score": per_school_math_df,
    "Average Reading Score" : per_school_reading_df,
    "% Passing Math" : school_students_passing_math_df,
    "% Passing Reading" : school_students_passing_reading_df,
    "% Overall Passing" : overall_passing_rate_df})

# Formatting
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.2f}".format)
#per_school_summary_df["Average Math Score"] = per_school_summary_df ["Average Math Score"].map("{:.0f}%".format)
#per_school_summary_df["Average Reading Score"] = per_school_summary_df ["Average Reading Score"].map("{:.2f}".format)
#per_school_summary_df["% Passing Reading"] = per_school_summary_df ["% Passing Reading"].map("{:.0f}%".format)

                                                                                                                                                                
print()
# Display the DataFrame
per_school_summary_df
```

# Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary_df.copy()
per_school_summary_df.copy().head()
```

```python
school_spending_df["Spending Ranges (Per Student)"]= pd.cut (per_school_capita_df,spending_bins, labels=labels, right = False)
school_spending_df.head()
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores_df = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores_df = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math_df = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading_df = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending_df = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
spending_math_scores_df.head()
```

```python
# Assemble into DataFrame
spending_summary_df = pd.DataFrame({
     "Average Reading Score" : spending_reading_scores_df,
    "Average Reading Score" :spending_reading_scores_df,
    "% Passing Math" : spending_passing_math_df,
    "% Passing Reading": spending_passing_reading_df,
    "% Overall Passing" :overall_passing_spending_df
    
})

# Display results
spending_summary_df
```
